# Part 2: Simplifying to a Binary Classification Problem (`Leak` vs. `NoLeak`)

Our first model struggled with the 4-class problem, which was likely caused by severe overfitting on a small dataset. As a next step, we will simplify the problem to a binary classification.

- **Class 1: `Leak`** (combining `Leak-Metal` and `Leak-NonMetal`)
- **Class 2: `NoLeak`** (combining `NoLeak-Metal` and `NoLeak-NonMetal`)

This doubles the number of examples per class, which should help the model learn more generalizable features and combat overfitting. We will reuse the spectrograms we've already loaded.


## 2.1 Prepare Binary Labels and Split Data


```python
# The spectrogram data `X_reshaped` is the same, so we can reuse it.
# We just need to create new binary labels. The original labels `y` are still in memory.

# Corrected logic: Check if the label starts with 'NoLeak'
y_binary = np.array(['NoLeak' if label.startswith('NoLeak') else 'Leak' for label in y])

# Encode the new binary labels ('NoLeak' -> 0, 'Leak' -> 1)
binary_label_encoder = LabelEncoder()
y_binary_encoded = binary_label_encoder.fit_transform(y_binary)

# Sanity check the new classes
print(f"Original classes: {np.unique(y)}")
print(f"New binary classes: {np.unique(y_binary)}")
print(f"Encoded binary labels: {np.unique(y_binary_encoded)}")
print(f"Class mapping: NoLeak -> {binary_label_encoder.transform(['NoLeak'])[0]}, Leak -> {binary_label_encoder.transform(['Leak'])[0]}")


# Split the dataset with the new binary labels
X_train_bin, X_val_bin, y_train_bin, y_val_bin = train_test_split(
    X_reshaped, y_binary_encoded, test_size=0.2, random_state=42, stratify=y_binary_encoded
)

print(f"\nShape of binary training data: {X_train_bin.shape}")
print(f"Shape of binary validation data: {X_val_bin.shape}")
```

## 2.2 Build and Train the Binary CNN Model

We will use a very similar CNN architecture as before, but with two crucial changes for binary classification:
1.  The final `Dense` layer will have only **1 neuron** instead of `NUM_CLASSES`.
2.  The activation function for this final layer will be **`sigmoid`**, which outputs a single probability value between 0 and 1.
3.  We will compile the model with **`binary_crossentropy`** loss, which is the standard loss function for two-class problems.


```python
binary_model = Sequential([
    # We use the same INPUT_SHAPE as before
    Conv2D(32, (3, 3), activation='relu', input_shape=INPUT_SHAPE),
    MaxPooling2D((2, 2)),
    
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),

    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    
    # --- Final layer for binary classification ---
    Dense(1, activation='sigmoid')
])

# Compile the model for binary classification
binary_model.compile(optimizer='adam',
                     loss='binary_crossentropy',
                     metrics=['accuracy'])

binary_model.summary()
```

```python
EPOCHS = 50
BATCH_SIZE = 32

binary_history = binary_model.fit(X_train_bin, y_train_bin,
                                epochs=EPOCHS,
                                batch_size=BATCH_SIZE,
                                validation_data=(X_val_bin, y_val_bin),
                                verbose=1)
```

## 2.3 Evaluate the Binary Model

Let's see how our new, simpler model performed. We'll plot the history and generate the reports just as we did before. Because the model outputs a single probability, we'll classify any output greater than 0.5 as the positive class (`Leak`) and less than or equal to 0.5 as the negative class (`NoLeak`).


```python
# Plotting training history for the binary model
pd.DataFrame(binary_history.history).plot(figsize=(12, 6))
plt.grid(True)
plt.gca().set_ylim(0, 1)
plt.title('Binary Model Training History')
plt.xlabel('Epoch')
plt.ylabel('Metric')
plt.show()

# Make predictions
y_pred_probs_bin = binary_model.predict(X_val_bin)
y_pred_bin = (y_pred_probs_bin > 0.5).astype("int32").reshape(-1)

# Generate and print the classification report
print("\\nBinary Classification Report:\\n")
print(classification_report(y_val_bin, y_pred_bin, target_names=binary_label_encoder.classes_))

# Generate and plot the confusion matrix
conf_matrix_bin = confusion_matrix(y_val_bin, y_pred_bin)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_bin, annot=True, fmt='d',
            xticklabels=binary_label_encoder.classes_,
            yticklabels=binary_label_encoder.classes_,
            cmap='Blues')
plt.title('Binary Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

# Part 3: Improving the Model with Data Augmentation

The results from our binary model were promising but showed clear signs of overfitting: the training accuracy was high, but the validation accuracy was much lower. This is a classic symptom of a model that has memorized the training data because the dataset is too small.

The best way to combat overfitting is to train on more data. Since we don't have more unique recordings, we'll use **data augmentation** to create artificial variations of our existing audio.

Our strategy:
1.  Define functions to perform a few types of audio augmentation: adding noise, time stretching, and pitch shifting.
2.  Create a new, larger training dataset. For each original audio clip in our training set, we will add the original version plus a few randomly augmented versions.
3.  **Crucially, we will not augment the validation data.** The validation set must remain a consistent, untouched benchmark to fairly evaluate our model's performance.
4.  Train a new model on this augmented dataset and compare the results.


## 3.1 Define Augmentation Functions and Re-create Dataset

First, we define the functions that will alter our audio data. Then, we will create our new dataset.

The correct methodology is to first split our *file paths* into training and validation sets. This ensures that no augmented versions of validation clips can ever leak into the training set. We will then process these two sets separately:
- **Training Set:** For each audio file, we'll create one original spectrogram and a few augmented versions.
- **Validation Set:** We'll only create one original spectrogram for each file.


```python
def add_noise(audio, noise_factor=0.005):
    noise = np.random.randn(len(audio))
    augmented_audio = audio + noise_factor * noise
    return augmented_audio

def time_stretch(audio, stretch_rate=0.8):
    return librosa.effects.time_stretch(y=audio, rate=stretch_rate)

def pitch_shift(audio, sample_rate, n_steps=4):
    return librosa.effects.pitch_shift(y=audio, sr=sample_rate, n_steps=n_steps)

# --- 1. Get all file paths and their binary labels ---
all_files = []
all_binary_labels = []
for label in classes:
    class_path = os.path.join(BASE_AUDIO_PATH, label)
    files = [os.path.join(class_path, f) for f in os.listdir(class_path) if f.endswith('.wav')]
    all_files.extend(files)
    
    binary_label = 'NoLeak' if label.startswith('NoLeak') else 'Leak'
    all_binary_labels.extend([binary_label] * len(files))

# --- 2. Split file paths into training and validation sets ---
train_files, val_files, train_labels, val_labels = train_test_split(
    all_files, all_binary_labels, test_size=0.2, random_state=42, stratify=all_binary_labels
)

print(f"Total files: {len(all_files)}")
print(f"Training files: {len(train_files)}")
print(f"Validation files: {len(val_files)}")

```

```python
# --- 3. Process the training and validation sets ---
X_train_aug, y_train_aug = [], []
X_val_aug, y_val_aug = [], []

print("Processing and augmenting training data...")
for file, label in tqdm(zip(train_files, train_labels), total=len(train_files)):
    try:
        audio, sr = librosa.load(file, sr=SAMPLE_RATE, res_type='kaiser_fast')

        # --- Original Audio ---
        spec_original = audio_to_melspectrogram(file)
        if spec_original is not None:
            X_train_aug.append(spec_original)
            y_train_aug.append(label)
        
        # --- Augmented Audio: Noise ---
        audio_noisy = add_noise(audio)
        # We need to pad/truncate the augmented audio just like the original
        if len(audio_noisy) < FIXED_LENGTH:
            audio_noisy = np.pad(audio_noisy, (0, FIXED_LENGTH - len(audio_noisy)), 'constant')
        else:
            audio_noisy = audio_noisy[:FIXED_LENGTH]
        spec_noisy = librosa.feature.melspectrogram(y=audio_noisy, sr=sr, n_mels=128)
        X_train_aug.append(librosa.power_to_db(spec_noisy, ref=np.max))
        y_train_aug.append(label)

        # --- Augmented Audio: Pitch Shift ---
        audio_pitched = pitch_shift(audio, sr)
        if len(audio_pitched) < FIXED_LENGTH:
            audio_pitched = np.pad(audio_pitched, (0, FIXED_LENGTH - len(audio_pitched)), 'constant')
        else:
            audio_pitched = audio_pitched[:FIXED_LENGTH]
        spec_pitched = librosa.feature.melspectrogram(y=audio_pitched, sr=sr, n_mels=128)
        X_train_aug.append(librosa.power_to_db(spec_pitched, ref=np.max))
        y_train_aug.append(label)

    except Exception as e:
        print(f"Error during augmentation for {file}: {e}")


print("\nProcessing validation data...")
for file, label in tqdm(zip(val_files, val_labels), total=len(val_files)):
    spec = audio_to_melspectrogram(file)
    if spec is not None:
        X_val_aug.append(spec)
        y_val_aug.append(label)

# --- 4. Convert to numpy arrays and prepare for model ---
X_train_aug = np.array(X_train_aug)[..., np.newaxis]
y_train_aug = np.array(y_train_aug)

X_val_aug = np.array(X_val_aug)[..., np.newaxis]
y_val_aug = np.array(y_val_aug)

# Encode labels
aug_label_encoder = LabelEncoder()
y_train_aug_enc = aug_label_encoder.fit_transform(y_train_aug)
y_val_aug_enc = aug_label_encoder.transform(y_val_aug)


print(f"\nNew training set size: {X_train_aug.shape[0]} samples")
print(f"New validation set size: {X_val_aug.shape[0]} samples")
```

## 3.2 Build and Train the Model on Augmented Data

Now, we'll define and train a new model. The architecture will be identical to our previous binary model, but this time it will be trained on our larger, augmented dataset. We will also increase the number of epochs slightly to allow the model more time to learn from the expanded data.


```python
aug_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=INPUT_SHAPE),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

aug_model.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

# We can train for more epochs since we have a larger dataset
EPOCHS_AUG = 50 
BATCH_SIZE = 32

aug_history = aug_model.fit(X_train_aug, y_train_aug_enc,
                            epochs=EPOCHS_AUG,
                            batch_size=BATCH_SIZE,
                            validation_data=(X_val_aug, y_val_aug_enc),
                            verbose=1)
```

## 3.3 Evaluate the Augmented Model

Finally, let's plot the results from our new model trained on the augmented data. We are hoping to see that the gap between the training accuracy and the validation accuracy has narrowed, which would indicate that the model is overfitting less and generalizing better.


```python
# Plotting training history for the augmented model
pd.DataFrame(aug_history.history).plot(figsize=(12, 6))
plt.grid(True)
plt.gca().set_ylim(0, 1)
plt.title('Augmented Model Training History')
plt.xlabel('Epoch')
plt.ylabel('Metric')
plt.show()

# Make predictions
y_pred_probs_aug = aug_model.predict(X_val_aug)
y_pred_aug = (y_pred_probs_aug > 0.5).astype("int32").reshape(-1)

# Generate and print the classification report
print("\\nAugmented Model Classification Report:\\n")
print(classification_report(y_val_aug_enc, y_pred_aug, target_names=aug_label_encoder.classes_))

# Generate and plot the confusion matrix
conf_matrix_aug = confusion_matrix(y_val_aug_enc, y_pred_aug)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_aug, annot=True, fmt='d',
            xticklabels=aug_label_encoder.classes_,
            yticklabels=aug_label_encoder.classes_,
            cmap='Blues')
plt.title('Augmented Model Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```
